# Exercício 01

Um fabricante de calçados afirma que os atletas podem aumentar a altura de seus saltos verticais usando o calçado de treinamento do fabricante.

As alturas dos saltos verticais de oito atletas aleatoriamente selecionados são medidas. Após usarem os calçados por 8 meses, suas alturas nos saltos verticais são novamente medidas. As alturas dos saltos
verticais (em polegadas) para cada atleta estão na tabela a seguir.

Aula 20-01.png

Para $\alpha=0,10$, há evidência suficiente para aceitar a afirmação do fabricante? Suponha que as alturas dos saltos verticais são normalmente distribuídas. (Adaptado de: *Coaches Sports Publishing*)

```python
from scipy import stats as st
import pandas as pd
# Conjunto de dados
saltos = pd.DataFrame({'Antes': [24, 22, 25, 28, 35, 32, 30, 27],
                       'Depois': [26, 25, 25, 29, 33, 34, 35, 30]})
display(saltos)
# d = (altura do salto antes do calçado) – (altura do salto após o calçado)
# H0: mu_d >= 0
# Ha: mu_d < 0 (afirmação)
tpadrao, p_valor = st.ttest_rel(a=saltos['Antes'], b=saltos['Depois'],
                                alternative='less')
print('*** Teste de amostras emparelhadas ***')
alpha = 0.10
if p_valor <= alpha:
  print(f'Hipótese nula rejeitada ao nível de significância de {alpha*100:.2f}%')
else:
  print(f'Hipótese nula NÃO rejeitada ao nível de significância de {alpha*100:.2f}%')
```

# Exercício 02

Um fabricante de calçados afirma que os atletas podem diminuir seus tempos na corrida de 40 jardas, usando o calçado de treinamento que ele produz.

Os tempos na corrida de 40 jardas de 12 atletas aleatoriamente selecionados são medidos. Após usarem os calçados por 8 meses, os seus tempos na corrida de 40 jardas são novamente medidos. Os tempos (em segundos) estão na tabela a seguir.

Aula 20-02.png

Com $\alpha= 0,05$, há evidência suficiente para aceitar a afirmação do fabricante? Suponha que os tempos são normalmente distribuídos. (Adaptado de: *Coaches Sports Publishing*)

```python
# Conjunto de dados
tempos = pd.DataFrame({'Antes': [4.85, 4.90, 5.08, 4.72, 4.62, 4.54, 5.25,
                                 5.18, 4.81, 4.57, 4.63, 4.77],
                       'Depois': [4.78, 4.90, 5.05, 4.65, 4.64, 4.50, 5.24,
                                  5.27, 4.75, 4.43, 4.61, 4.82]})
display(tempos)
# d = (tempo de 40yd antes do calçado) – (tempo de 40yd após o calçado)
# H0: mu_d <= 0
# Ha: mu_d > 0 (afirmação)
tpadrao, p_valor = st.ttest_rel(a=tempos['Antes'], b=tempos['Depois'],
                                alternative='greater')
print('*** Teste de amostras emparelhadas ***')
alpha = 0.05
if p_valor <= alpha:
  print(f'Hipótese nula rejeitada ao nível de significância de {alpha*100:.2f}%')
else:
  print(f'Hipótese nula NÃO rejeitada ao nível de significância de {alpha*100:.2f}%')
```

# Exercício 03

A equipe de campanha de um deputado estadual quer determinar se a avaliação do desempenho do deputado (0 – 100) mudou do ano passado para o atual. A tabela a seguir mostra as avaliações do desempenho do
deputado, dos mesmos 16 eleitores selecionados aleatoriamente, no ano passado e no atual.

Aula 20-03.png

Para $\alpha=0,01$, há evidência suficiente para concluir que a avaliação do desempenho do deputado mudou? Suponha que as avaliações do desempenho são normalmente distribuídas.

```python
# Conjunto de dados
notas = pd.DataFrame({'Ano passado': [60, 54, 78, 84, 91, 25, 50, 65,
                                      68, 81, 75, 45, 62, 79, 58, 63],
                      'Ano atual': [56, 48, 70, 60, 85, 40, 40, 55,
                                    80, 75, 78, 50, 50, 85, 53, 60]})
display(notas)
# d = (avaliação no ano passado) – (avaliação no ano atual)
# H0: mu_d = 0
# Ha: mu_d != 0 (afirmação)
tpadrao, p_valor = st.ttest_rel(a=notas['Ano passado'], b=notas['Ano atual'],
                                alternative='two-sided')
print('*** Teste de amostras emparelhadas ***')
alpha = 0.01
if p_valor <= alpha:
  print(f'Hipótese nula rejeitada ao nível de significância de {alpha*100:.2f}%')
else:
  print(f'Hipótese nula NÃO rejeitada ao nível de significância de {alpha*100:.2f}%')
```

# Exercício 04

Um médico pesquisador quer determinar se uma droga altera a temperatura do corpo. Sete indivíduos são selecionados aleatoriamente como cobaias, e a temperatura do corpo (em graus Fahrenheit) de cada
um é medida. A droga é então dada aos indivíduos e, após 20 minutos, a temperatura do corpo de cada um é medida novamente. Os resultados
estão listados na tabela a seguir.

Aula 20-04.png

Com $\alpha = 0,05$, há evidência suficiente para
concluir que a droga altera a temperatura do corpo? Suponha que as temperaturas do corpo são distribuídas normalmente.

```python
# Conjunto de dados
temp = pd.DataFrame({'Ti': [101.8, 98.5, 98.1, 99.4, 98.9, 100.2, 97.9],
                     'Tf': [99.2, 98.4, 98.2, 99.0, 98.6, 99.7, 97.8]})
display(temp)
# d = (temperatura antes da droga) – (temperatura após a droga)
# H0: mu_d = 0
# Ha: mu_d != 0 (afirmação)
tpadrao, p_valor = st.ttest_rel(a=temp['Ti'], b=temp['Tf'],
                                alternative='two-sided')
print('*** Teste de amostras emparelhadas ***')
alpha = 0.05
if p_valor <= alpha:
  print(f'Hipótese nula rejeitada ao nível de significância de {alpha*100:.2f}%')
else:
  print(f'Hipótese nula NÃO rejeitada ao nível de significância de {alpha*100:.2f}%')
```
