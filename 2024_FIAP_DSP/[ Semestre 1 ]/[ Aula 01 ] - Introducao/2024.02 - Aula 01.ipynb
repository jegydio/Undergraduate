## Setup Python

```python
import matplotlib.pyplot as plt  # Usado para criar gráficos e visualizações
from sympy import symbols, diff, solve, Eq, lambdify, integrate  # sympy é uma biblioteca para matemática simbólica
import numpy as np  # Numpy é uma biblioteca fundamental para operações matemáticas e manipulação de arrays
```

`Sympy` é uma biblioteca Python para computação simbólica. Ela oferece uma ampla gama de funcionalidades para manipulação de expressões matemáticas simbólicas, tornando-a uma ferramenta poderosa para matemática simbólica, álgebra, cálculo e muito mais. Aqui estão alguns aspectos chave da biblioteca `sympy`:

1. **Manipulação de Expressões Simbólicas:** Sympy permite a definição e manipulação de expressões matemáticas simbólicas. Isso significa que, ao contrário de cálculos numéricos que trabalham com valores aproximados, o sympy trabalha com os símbolos matemáticos propriamente ditos. Por exemplo, pode-se manipular a expressão $ x^2 + y^2 $ diretamente.

2. **Cálculo Diferencial e Integral:** Uma das principais funcionalidades do sympy é a capacidade de realizar cálculo diferencial e integral. Isso inclui derivar e integrar expressões simbólicas, limites, expansões em séries e muito mais.

3. **Solução de Equações:** Sympy pode ser usado para resolver equações e sistemas de equações. Isso não se limita apenas a equações algébricas, mas também a equações diferenciais.

4. **Álgebra Linear:** A biblioteca oferece suporte a operações de álgebra linear, incluindo manipulação de matrizes, determinantes, inversas, autovalores e autovetores.

5. **Simplificação e Expansão:** Sympy pode simplificar expressões matemáticas complexas, expandi-las ou fatorá-las, o que é útil para tornar os cálculos mais tratáveis ou para fins de prova e demonstração.

6. **Transformações e Funções Especiais:** A biblioteca inclui suporte para várias transformações (como a Transformada de Laplace) e uma variedade de funções especiais e constantes matemáticas.

7. **Criação de Código:** Sympy pode gerar código numérico a partir de expressões simbólicas para várias linguagens de programação, incluindo C, Fortran, e Python (através da função `lambdify`).

8. **Interface Amigável:** Sympy trabalha bem com Jupyter Notebooks, oferecendo uma bela formatação de saída LaTeX para as expressões matemáticas.

9. **Extensível e Integrável:** É possível estender o sympy com suas próprias funções personalizadas, e ele pode ser integrado com outras bibliotecas Python, como NumPy, Matplotlib, e Pandas para análises mais complexas.

Em resumo, Sympy é uma ferramenta versátil e poderosa para matemáticos, cientistas, engenheiros e qualquer pessoa que precise realizar cálculos matemáticos simbólicos em Python.

## Definindo os parâmetros do problema #1

```python
# Definindo 'x' como uma variável simbólica para uso em funções matemáticas
x = symbols('x')  # x representa o número de aparelhos vendidos por semana

preco_base = 350  # Preço base de um videogame sem desconto
desconto_unitario = 10  # Valor do desconto por unidade
aumento_unidades = 20  # Aumento no número de unidades vendidas semanais por desconto aplicado
unidades_base = 200  # Unidades vendidas sem desconto
```

## Funções

```python
# Função demanda: p(x) = 350 - (10/20)*(x - 200)
demanda = preco_base - (desconto_unitario/aumento_unidades) * (x - unidades_base)
demanda

# Função receita: R(x) = x * demanda
receita = x * demanda

# Derivada da função receita em relação a x
derivada_receita = diff(receita, x)
derivada_receita
```

```python
# Encontrar o valor de x que maximiza a receita (onde a derivada é zero)
ponto_maximo = solve(Eq(derivada_receita, 0), x)  # Isso extrai o primeiro valor da lista
ponto_maximo
```

1. **`solve(Eq(derivada_receita, 0), x)`**:
   
   - `solve(...)`: Esta é uma função da biblioteca `sympy` usada para resolver equações. Ela pode ser usada para encontrar as raízes de uma equação, ou seja, os valores de `x` que fazem a equação ser verdadeira.
   - `Eq(derivada_receita, 0)`: Aqui, a função `Eq` é usada para criar uma equação. O primeiro argumento (`derivada_receita`) é a derivada da função de receita em relação à variável `x`. O segundo argumento (`0`) indica que estamos igualando esta derivada a zero. Em termos matemáticos, encontrar os pontos onde a derivada de uma função é igual a zero é uma forma comum de encontrar os pontos de máximo ou mínimo dessa função.
   - `x`: Este é o símbolo para o qual estamos resolvendo a equação.

2. **`ponto_maximo = ...`**: Esta linha armazena o resultado da função `solve` na variável `ponto_maximo`. O resultado pode ser:
   - Um único valor: indicando que há um único ponto de máximo (ou mínimo).
   - Uma lista de valores: indicando que há múltiplos pontos de máximo (ou mínimo).
   - Vazio: indicando que não há pontos de máximo (ou mínimo) reais.


```python
# Função demanda: p(ponto_maximo)
preco = preco_base - (desconto_unitario/aumento_unidades) * (ponto_maximo[0] - unidades_base)
preco
```

```python
desconto = preco_base - preco
desconto
```

# Análise Gráfica

```python
# Definindo um intervalo de x mais detalhado para uma curva mais suave
x_vals_fine = np.linspace(0, 1000, 5000)  # Aumentando o número de pontos
receita_func = lambdify(x, receita, modules=['numpy'])
receita_vals_fine = receita_func(x_vals_fine)
ponto_maximo = solve(Eq(derivada_receita, 0), x)[0]  # Isso extrai o primeiro valor da lista
receita_maxima = receita_func(ponto_maximo)

# Plotando a relação entre o número de unidades vendidas e a receita com melhorias visuais
plt.figure(figsize=(10, 6))
plt.plot(x_vals_fine, receita_vals_fine, label='Receita em função de x', color='blue', linewidth=2)
plt.scatter(ponto_maximo, receita_maxima, color='red', s=50)  # Marcando o ponto máximo com um tamanho maior
plt.text(ponto_maximo, receita_maxima, f' Máximo em x={ponto_maximo}', verticalalignment='bottom', color='darkred')

# Enfatizando os eixos de referência x e y
plt.axhline(0, color='black', linewidth=1.3)  # Linha horizontal em y=0
plt.axvline(0, color='black', linewidth=1.3)  # Linha vertical em x=0

# Adicionando títulos e etiquetas
plt.title('Gráfico da Função Receita', fontsize=16, fontweight='bold')
plt.xlabel('Número de Unidades Vendidas', fontsize=14)
plt.ylabel('Receita (R$)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()  # Ajusta automaticamente os subplots para caber na área de figura
plt.show()
```

## Definindo os parâmetros para o problema #2

```python
# Definição da variável simbólica
x = symbols('x')

# Definição da função f(x)
f_x = 0.006*x*(10-x)
```

```python
# a. Verificando se f(x) é uma função densidade de probabilidade

# Uma função é uma densidade de probabilidade se a integral de f(x) de -inf a +inf é igual a 1
# Neste caso, precisamos integrar de 0 a 10
integral_f_x = integrate(f_x, (x, 0, 10))
integral_f_x
```

```python
# b. Encontrando P(4 <= x <= 8)
probabilidade = integrate(f_x, (x, 4, 8))
probabilidade
```
