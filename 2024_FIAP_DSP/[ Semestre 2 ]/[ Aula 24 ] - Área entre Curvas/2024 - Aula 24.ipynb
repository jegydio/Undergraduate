```python
# Refazendo o clculo para verificar a rea corretamente

# Passo 1: Encontrar os pontos corretos de interseo entre f(y) e g(y)
pontos_intersecao = solve(Eq(f_y, g_y), y)
pontos_intersecao = sorted([float(p) for p in pontos_intersecao])  # Ordenando

# Passo 2: Calcular a rea em cada intervalo
# Observao: A rea ser a integral da diferena entre as funes em cada sub-regio.

def calcular_area_revisada(inicio, fim):
    """Calcula a rea entre duas funes no intervalo dado."""
    area = integrate(abs(g_y - f_y), (y, inicio, fim))
    return abs(float(area))

# Clculo da rea total (separando em cada parte)
area1 = calcular_area_revisada(pontos_intersecao[0], pontos_intersecao[1])
area2 = calcular_area_revisada(pontos_intersecao[1], pontos_intersecao[2])
area_total = area1 + area2

# Preparando os valores para os dois intervalos: de y = -2 a 0 e de y = 0 a 2
y_vals1 = np.linspace(-2, 0, 200)  # Intervalo 1: -2 a 0
y_vals2 = np.linspace(0, 2, 200)   # Intervalo 2: 0 a 2

# Calculando os valores das funes nos dois intervalos
f_vals1 = y_vals1**3 - y_vals1  # f(y) para o intervalo 1
g_vals1 = 3 * y_vals1           # g(y) para o intervalo 1

f_vals2 = y_vals2**3 - y_vals2  # f(y) para o intervalo 2
g_vals2 = 3 * y_vals2           # g(y) para o intervalo 2

# Criando o grfico com destaque para as duas reas
plt.figure(figsize=(8, 6))

# Primeira rea: de y = -2 a 0
plt.plot(f_vals1, y_vals1, label='$f(y) = y^3 - y$', color='orange')
plt.plot(g_vals1, y_vals1, label='$g(y) = 3y$', color='red')
plt.fill_betweenx(y_vals1, f_vals1, g_vals1, where=(g_vals1 <= f_vals1), color='lightgray', alpha=0.5)

# Segunda rea: de y = 0 a 2
plt.plot(f_vals2, y_vals2, color='orange')
plt.plot(g_vals2, y_vals2, color='red')
plt.fill_betweenx(y_vals2, f_vals2, g_vals2, where=(g_vals2 >= f_vals2), color='lightgray', alpha=0.5)


# Adicionando detalhes no grfico
plt.title('rea completa entre as curvas $f(y) = y^3 - y$ e $g(y) = 3y$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.ylim(-2.5, 2.5)  # Definindo limites adequados para o eixo y
plt.legend(loc='upper left')
plt.grid(True)

# Exibindo o grfico ajustado
plt.show()
# Exibir a rea total corrigida
area_total
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve, integrate, cos, sin, pi

# ---- Parte (a): y = 2x^2 + 10 e y = 4x + 16 ----

# Definio das funes em SymPy
x = symbols('x')
f_a = 2 * x**2 + 10
g_a = 4 * x + 16

# Encontrando pontos de interseo
intersecao_a = solve(Eq(f_a, g_a), x)
intersecao_a = sorted([float(p) for p in intersecao_a])

# Calculando a rea entre as curvas (f_a - g_a)
area_a = integrate(abs(f_a - g_a), (x, intersecao_a[0], intersecao_a[1]))

# Preparando dados para o grfico
x_vals_a = np.linspace(intersecao_a[0] - 1, intersecao_a[1] + 1, 400)
f_vals_a = 2 * x_vals_a**2 + 10
g_vals_a = 4 * x_vals_a + 16

# Criando o grfico para a parte (a)
plt.figure(figsize=(8, 6))
plt.plot(x_vals_a, f_vals_a, label=r'$y = 2x^2 + 10$', color='blue')
plt.plot(x_vals_a, g_vals_a, label=r'$y = 4x + 16$', color='red')
plt.fill_between(x_vals_a, f_vals_a, g_vals_a, where=(f_vals_a <= g_vals_a), color='lightgray', alpha=0.5)

plt.title('rea entre $y = 2x^2 + 10$ e $y = 4x + 16$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)

# Exibir o grfico para a parte (a)
plt.show()

# Exibir a rea calculada para a parte (a)
area_a
```

```python
# ---- Parte (b): y = sin(x) e y = cos(x), no intervalo [0, pi/2] ----

# Definio das funes em SymPy
f_b = sin(x)
g_b = cos(x)

# Calculando a rea entre as curvas no intervalo [0, pi/2]
area_b = integrate(abs(f_b - g_b), (x, 0, pi/2))

# Preparando dados para o grfico
x_vals_b = np.linspace(0, np.pi / 2, 800)
f_vals_b = np.sin(x_vals_b)
g_vals_b = np.cos(x_vals_b)

# Criando o grfico para a parte (b)
plt.figure(figsize=(8, 6))
plt.plot(x_vals_b, f_vals_b, label=r'$y = \sin(x)$', color='blue')
plt.plot(x_vals_b, g_vals_b, label=r'$y = \cos(x)$', color='red')
plt.fill_between(x_vals_b, f_vals_b, g_vals_b, where=(f_vals_b >= g_vals_b), color='lightgray', alpha=0.5)
plt.fill_between(x_vals_b, f_vals_b, g_vals_b, where=(f_vals_b <= g_vals_b), color='lightgray', alpha=0.5)

plt.title('rea entre $y = \sin(x)$ e $y = \cos(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)

# Exibir o grfico para a parte (b)
plt.show()

# Exibir a rea calculada para a parte (b)
area_b
```

```python
# ---- Parte (c): x = y^2/2 - 3 e y = x - 1 ----

# Definindo as funes em SymPy
y = symbols('y')
f_c = (y**2 / 2) - 3  # x = y^2/2 - 3
g_c = y + 1  # y = x - 1 --> x = y + 1

# Encontrando os pontos de interseo
intersecao_c = solve(Eq(f_c, g_c), y)
intersecao_c = sorted([float(p) for p in intersecao_c])

# Calculando a rea entre as curvas no intervalo dos pontos de interseo
area_c = integrate(abs(f_c - g_c), (y, intersecao_c[0], intersecao_c[1]))

# Preparando os dados para o grfico
y_vals_c = np.linspace(intersecao_c[0] - 1, intersecao_c[1] + 1, 400)
f_vals_c = (y_vals_c**2 / 2) - 3
g_vals_c = y_vals_c + 1

# Criando o grfico para a parte (c)
plt.figure(figsize=(8, 6))
plt.plot(f_vals_c, y_vals_c, label=r'$x = \frac{y^2}{2} - 3$', color='blue')
plt.plot(g_vals_c, y_vals_c, label=r'$y = x - 1$', color='red')
plt.fill_betweenx(y_vals_c, f_vals_c, g_vals_c, where=(f_vals_c <= g_vals_c), color='lightgray', alpha=0.5)
plt.title(r'rea entre $x = \frac{y^2}{2} - 3$ e $y = x - 1$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend(loc='upper left')
plt.grid(True)

# Exibir o grfico para a parte (c)
plt.show()

# Exibir a rea calculada para a parte (c)
area_c
```
