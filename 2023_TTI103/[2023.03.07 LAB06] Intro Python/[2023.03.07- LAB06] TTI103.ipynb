```python
# Entrada de dados
a = float(input('Entre o comprimento "a": '))
b = float(input('Entre o comprimento "b": '))
c = float(input('Entre o comprimento "c": '))

# Verificação da condição de existência do triângulo
if (abs(b - c) < a < abs(b + c)) and (abs(a - c) < b < abs(a + c)) and (abs(a - b) < c < (a + b)):
  if a == b == c:
    print('\nOs lados a, b e c formam um triânculo do tipo Equilátero')
  elif a != b != c:
      print('\nOs lados a, b e c formam um triânculo do tipo Escaleno')
  else:
      print('\nOs lados a, b e c formam um triânculo do tipo Isósceles')
else:
  print('\nOs lados a, b e c não formam um triânculo!')
```

```python
# Entrada de dados
massa = float(input('Entre com o seu peso [kg]: '))
altura = float(input('Entre a sua altura [m] "b": '))

# Cálculo o IMC
IMC = massa/altura**2

# Classificação:
if IMC < 18.5:
  print(f'\n IMC < {IMC:.1f} - Abaixo do Peso')
elif 18.5 <= IMC < 25:
  print(f'\n 18.5 <= IMC {IMC:.1f} < 25 - Saudável')
elif 25 <= IMC < 30:
  print(f'\n 25 <= IMC {IMC:.1f} < 30 - Sobrepeso')
elif 30 <= IMC < 35:
  print(f'\n 30 <= IMC {IMC:.1f} < 35 - Obesidade Grau I')
elif 35 <= IMC < 40:
  print(f'\n 35 <= IMC {IMC:.1f} < 40 - Obesidade Grau II')
else:
  print(f'\n IMC {IMC:.1f} >= 40 - Obesidade Grau III')
```
