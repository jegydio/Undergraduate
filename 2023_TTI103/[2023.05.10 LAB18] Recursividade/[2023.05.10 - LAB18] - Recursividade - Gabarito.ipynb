**EX01 - Multiplicação Recursiva**: Elabore uma função recursiva que recebe dois parâmetros `x` e `y` (assuma valores maiores ou iguais a zero). A função deve retornar o valor de `x` multiplicado por `y`. 

Lembre-se que a multiplicação pode ser realizada por meio de adições sequenciais. 

Veja um exemplo: 

 `7 × 4 = 4 + 4 + 4 + 4 + 4 + 4 + 4 = 28`

```python
# Código

# Define uma função para realizar a multiplicação de dois números inteiros 
# positivos de forma recursiva.
def mult_rec(x, y):
    # Verifica se algum dos argumentos é igual a zero, pois qualquer número 
    # multiplicado por zero é igual a zero.
    if x == 0 or y == 0:
        return 0
    # Retorna a soma do segundo argumento com a multiplicação do primeiro 
    # argumento decrementado em uma unidade e o segundo argumento.
    else:
        return y + mult_rec(x-1, y)

# Testes da função
print('-- Multiplicação recursiva --')
print(mult_rec(4, 6)) # 4 * 6 = 24
print(mult_rec(5, 1)) # 5 * 1 = 5
print(mult_rec(0, 456)) # 0 * 456 = 0
```

**EX02 – Linhas recursivas**: Escreva uma função recursiva que recebe um inteiro positivo `n` e exibe `n` linhas de asteriscos. 

A primeira linha deve exibir um asterisco; a segunda dois asteriscos e assim por diante até a `n-ésima` linha, com a exibição de `n` asteriscos.

```python
# Código

# Define uma função para imprimir uma linha com n asteriscos de forma recursiva.
def imprime_asteriscos(n):
    # Verifica se n é maior que 1, pois se for, é necessário chamar a função 
    # recursivamente com n-1.
    if n > 1:
        imprime_asteriscos(n-1)
    # Imprime uma linha com n asteriscos.
    print('*'*n)

# Testes da função
print('-- Impressão de asteriscos [recursivo] --')
imprime_asteriscos(10) # Imprime 10 linhas com um número crescente de asteriscos.
imprime_asteriscos(0) # Imprime uma linha sem asteriscos, já que o argumento é zero.
```

**EX03 – Máximo valor contido em uma lista:** Construa uma função que recebe uma lista de números como parâmetro e retorna o máximo valor nela contido. A função deve ser recursiva!

```python
# Código

# Define uma função para encontrar o valor máximo em uma lista de forma recursiva.
def max_valor(lista):
    # Verifica se a lista tem apenas um elemento, retornando o valor desse 
    #elemento caso afirmativo.
    if len(lista) == 1:
        return lista[0]
    else:
        # Obtém o primeiro elemento da lista e o valor máximo dos demais 
        # elementos da lista, chamando a função recursivamente com a lista a 
        # partir do segundo elemento.
        primeiro = lista[0]
        max_demais = max_valor(lista[1:])
        # Compara o valor máximo dos demais elementos com o primeiro elemento, 
        #retornando o maior dos dois.
        if max_demais > primeiro:
            return max_demais
        else:
            return primeiro

# Define uma outra função para encontrar o valor máximo em uma lista de forma 
#recursiva, com uma sintaxe diferente da função anterior.
def max_valor2(lista):
    # Verifica se a lista tem apenas um elemento, retornando o valor desse 
    #elemento caso afirmativo.
    if len(lista) == 1:
        return lista[0]
    else:
        # Obtém o valor máximo dos demais elementos da lista, chamando a função 
        # recursivamente com a lista a partir do segundo elemento, 
        # e o compara com o primeiro elemento usando a sintaxe ternária.
        max_demais = max_valor(lista[1:])
        return max_demais if max_demais > lista[0] else lista[0]

# Testes das funções
print('-- Máximo valor em uma lista [recursiva] --')
print(max_valor([-3, 4, 6, 90, 45])) # Imprime o valor máximo da lista: 90.
print(max_valor([-5])) # Imprime o único elemento da lista, que é -5.
print(max_valor2([-30, 54, 57, 190, 66])) # Imprime o valor máximo da lista: 190
print(max_valor2([-5])) # Imprime o único elemento da lista, que é -5
```

**EX04 – Soma recursiva dos elementos de uma lista:** Construa uma função que recebe uma lista de números e retorna a soma dos elementos da lista, calculada recursivamente.

```python
# Código

# =============================================================================
# Soma recursiva dos elementos de uma lista
# =============================================================================

# Define uma função para somar todos os elementos de uma lista de forma recursiva.
def soma_lista(lista):
    # Verifica se a lista está vazia, retornando 0 caso afirmativo.
    if len(lista) == 0:
        return 0
    else:
        # Retorna a soma do primeiro elemento da lista com a soma dos demais 
        # elementos da lista, chamando a função recursivamente com a lista a 
        # partir do segundo elemento.
        return lista[0] + soma_lista(lista[1:])

# Define uma outra função para somar todos os elementos de uma lista de 
# forma recursiva, com uma sintaxe diferente da função anterior.
def soma_lista2(lista):
    # Retorna 0 caso a lista esteja vazia, caso contrário retorna a soma do 
    # primeiro elemento da lista com a soma dos demais elementos da lista, 
    # chamando a função recursivamente com a lista a partir do segundo elemento, 
    # usando a sintaxe ternária.
    return 0 if len(lista) == 0 else lista[0] + soma_lista(lista[1:])

# Testes das funções
print('-- Soma recursiva de elementos em uma lista --')
print(soma_lista([])) # Imprime 0, pois a lista está vazia.
print(soma_lista([2])) # Imprime 2, pois a lista contém apenas um elemento.
print(soma_lista([-5, 0, 4, 6])) # Imprime 5, que é a soma de todos os elementos da lista.
print(soma_lista2([])) # Imprime 0, pois a lista está vazia, usando a segunda função.
print(soma_lista2([2])) # Imprime 2, pois a lista contém apenas um elemento, usando a segunda função.
print(soma_lista2([-5, 0, 4, 6])) # Imprime 5, que é a soma de todos os elementos da lista, usando a segunda função.
```

**EX05 – Soma dos n primeiros inteiros:** Elabore uma função que recebe um inteiro `n` (positivo) como um parâmetro e retorna a soma de todos os inteiros positivos até `n`. Por exemplo, se `n = 50`, a função deve retornar a soma `1 + 2 + 3 + ⋯ + 48 + 49 + 50`. Use recursão no cálculo da soma.

```python
# =============================================================================
# Soma recursiva de todos os inteiros até n (n >= 0)
# =============================================================================
# Define uma função para somar todos os inteiros de 1 até n de forma recursiva.
def soma_todos(n):
    # Retorna 1 se n for igual a 1, caso contrário retorna a soma de n com a 
    # chamada recursiva da função para n-1.
    return 1 if n == 1 else n + soma_todos(n-1)

# Define uma outra função para somar todos os inteiros de 1 até n de forma 
# recursiva, com uma estrutura de controle diferente da função anterior.
def soma_todos2(n):
    # Verifica se n é maior que 1, retornando n + soma_todos2(n-1) caso afirmativo.
    if n > 1:
        return n + soma_todos2(n-1)
    # Retorna 1 se n for igual a 1.
    elif n == 1:
        return 1
    # Lança uma exceção se n for negativo.
    else:
        raise Exception('Número negativo')

# Testes das funções
print('-- Soma recursiva de todos os inteiros até n --')
print(soma_todos(1)) # Imprime 1, pois a soma de todos os inteiros até 1 é 1.
print(soma_todos(10)) # Imprime 55, que é a soma de todos os inteiros até 10.
print(soma_todos2(50)) # Imprime 1275, que é a soma de todos os inteiros até 50, usando a segunda função.
```

**EX06 – Potência recursiva:** Escreva uma função que recebe dois inteiros `x` e `y`. A função deve usar a recursão para o cálculo de `x` elevado a `y`.

```python
# Código

# Define uma função para calcular a potência de um número de forma recursiva.
def potencia(x, y):
    # Verifica se y é igual a 0, retornando 1 caso afirmativo.
    if y == 0:
        return 1
    # Verifica se y é igual a 1, retornando x caso afirmativo.
    elif y == 1:
        return x
    else:
        # Retorna a multiplicação de x pela chamada recursiva da função com os argumentos x e y-1.
        return x * potencia(x, y-1)

# Testes da função
print('-- Potência recursiva --')
print(potencia(3, 0)) # Imprime 1, pois 3 elevado a 0 é 1.
print(potencia(4, 1)) # Imprime 4, pois 4 elevado a 1 é 4.
print(potencia(3, 4)) # Imprime 81, pois 3 elevado a 4 é 81.
```
