**EX01 - Multiplicação Recursiva**: Elabore uma função recursiva que recebe dois parâmetros `x` e `y` (assuma valores maiores ou iguais a zero). A função deve retornar o valor de `x` multiplicado por `y`. 

Lembre-se que a multiplicação pode ser realizada por meio de adições sequenciais. 

Veja um exemplo: 

`7 × 4 = 4 + 4 + 4 + 4 + 4 + 4 + 4 = 28`

```python
# Código

def mult_rec(x, y):
    if x == 0 or y == 0:
        return 0
    else:
        return y + mult_rec(x-1, y)

# Testes
print('Multiplicação recursiva:')
print(mult_rec(4, 6)) #Resp: 24
print(mult_rec(5, 1)) #Resp: 5
print(mult_rec(0, 500)) #Resp: 0
```

**EX02 – Linhas recursivas**: Escreva uma função recursiva que recebe um inteiro positivo `n` e exibe `n` linhas de asteriscos. 

A primeira linha deve exibir um asterisco; a segunda dois asteriscos e assim por diante até a `n-ésima` linha, com a exibição de `n` asteriscos.

```python
# Código

def imprime_asteriscos(n):
    if n > 1:
        imprime_asteriscos(n-1)
    print('*'*n)

print('Impressão de asteriscos recursivo:')
imprime_asteriscos(8)
imprime_asteriscos(0)
imprime_asteriscos(5)
```

**EX03 – Máximo valor contido em uma lista:** Construa uma função que recebe uma lista de números como parâmetro e retorna o máximo valor nela contido. A função deve ser recursiva!

```python
# Código

def max_valor(lista):
    if len(lista) == 1:
        return lista[0]
    else:
        primeiro = lista[0]
        max_demais = max_valor(lista[1:])
    if max_demais > primeiro:
        return max_demais
    else:
        return primeiro


print('Máximo valor em uma lista:')
print(max_valor([-3, 400, 6, 90, 500]))
print(max_valor([-3]))
```

**EX04 – Soma recursiva dos elementos de uma lista:** Construa uma função que recebe uma lista de números e retorna a soma dos elementos da lista, calculada recursivamente.

```python
# Código

def soma_lista(lista):
    if len(lista) == 0:
        return 0
    else:
        return lista[0] + soma_lista(lista[1:])

print('Soma recursiva de elementos em uma lista:')
print(soma_lista([]))
print(soma_lista([6]))
print(soma_lista([-5, 0, 4, 6]))
```

**EX05 – Soma dos n primeiros inteiros:** Elabore uma função que recebe um inteiro `n` (positivo) como um parâmetro e retorna a soma de todos os inteiros positivos até `n`. Por exemplo, se `n = 50`, a função deve retornar a soma `1 + 2 + 3 + ⋯ + 48 + 49 + 50`. Use recursão no cálculo da soma.

```python
# Código

def soma_todos(n):
  return 1 if n == 1 else n + soma_todos(n-1)

print('-- Soma recursiva de todos os inteiros até n --')
print(soma_todos(1)) # Imprime 1, pois a soma de todos os inteiros até 1 é 1.
print(soma_todos(10)) # Imprime 55, que é a soma de todos os inteiros até 10.
print(soma_todos(100)) # Imprime 1275, que é a soma de todos os inteiros até 50, usando a segunda função.
```
