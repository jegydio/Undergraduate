```python
# tabela de pre√ßos das frutas
preco_morango_ate_5 = 2.5
preco_morango_acima_5 = 2.2
preco_maca_ate_5 = 1.8
preco_maca_acima_5 = 1.5

# entrada de dados
morango = float(input('Peso dos morangos [kg]: '))
maca = float(input('Peso das ma√ß√£s [kg]:    '))

# c√°lculo do valor em fun√ß√£o do peso

if morango <= 5:
    valor_morango = preco_morango_ate_5*morango
else:
    valor_morango = preco_morango_acima_5*morango

print(f'\nPeso total de morangos:     {morango:.2f} kg')
print(f'Valor total de morangos:    R$ {valor_morango:.2f}')
 

if maca <= 5:
    valor_maca = preco_maca_ate_5*maca
else:
    valor_maca = preco_maca_acima_5*maca

print(f'\nPeso total de macas:        {maca:.2f} kg')
print(f'Valor total de macas:       R$ {valor_maca:.2f}')

# c√°lculo com os descontos

valor_subtotal = valor_maca + valor_morango
peso_total = maca + morango


if peso_total >= 8:
    desconto = 0.1*valor_subtotal
    valor_total = 0.9*valor_subtotal
elif valor_subtotal >= 25:
    desconto = 0.1*valor_subtotal
    valor_total = 0.9*valor_subtotal
else:
    desconto = 0
    valor_total = valor_subtotal

print(f'\nPeso total das compras:     {peso_total:.2f} kg')
print(f'Valor subtotal das compras: R$ {valor_subtotal:.2f}')
print(f'\nDescontos:                  R$ {desconto:.2f}')
print(f'Valor total das compras:    R$ {valor_total:.2f}')
```

```python
# usando o Pandas para manipula√ß√£o de tabelas no Python
import pandas as pd

# cria√ß√£o de tabela de usu√°rios
df = pd.DataFrame({
    'nome': ['Jo√£o', 'Maria', 'Pedro', 'Ana'],
    'codigo': ['23.10000-1', '23.10001-2', '23.10002-3', '23.10004-5'],
    'senha': ['0123', '1234', '2345', '3456']
})


codigo = input('C√≥digo do usu√°rio: ')

# buscar um valor em toda a tabela
#valor = codigo
encontrado = False

for coluna in df.columns:
    for valor_celula in df[coluna]:
        if codigo == valor_celula:
            encontrado = True
            print(f'C√≥digo do usu√°rio {codigo} √© v√°lido.')
            #break
    if encontrado:
        break

if not encontrado:
    print(f'C√≥digo do usu√°rio {codigo} √© inv√°lido.')
```
