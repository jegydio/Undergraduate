### Ex02

Neste exercício, crie uma rotina que, aleatoriamente, gera uma sequência de 500 numéros dentre 10000 possibilidades. A partir daí, encontre o menor, maior e a média dos números amazenados.

image.png

Resposta

```python
# Código Python versão 1
import random

# Parâmetros
num_aleatorios = 500
num_possibilidades = 10000

# Gerar sequência aleatória

numeros =[]

for i in range(num_aleatorios):
  numeros.append(random.randint(1, num_possibilidades))

# Encontrar o menor, maior e a média dos números
menor = min(numeros)
maior = max(numeros)
media = sum(numeros) / len(numeros)

# Exibir resultados

# Definir a largura das colunas
largura_col1 = 20
largura_col2 = 10

# Exibir cabeçalho
print(f"{'Descrição':<{largura_col1}}{'Valor':<{largura_col2}}")
print("-" * (largura_col1 + largura_col2))

# Exibir resultados
print(f"{'Menor número:':<{largura_col1}}{menor:<{largura_col2}}")
print(f"{'Maior número:':<{largura_col1}}{maior:<{largura_col2}}")
print(f"{'Média dos números:':<{largura_col1}}{media:<{largura_col2}.2f}")
```

```python
# Código Python versão 2
import random

# Parâmetros
num_aleatorios = 500
num_possibilidades = 10000

# Gerar sequência aleatória
numeros = [random.randint(1, num_possibilidades) for _ in range(num_aleatorios)]

# Encontrar o menor, maior e a média dos números
menor = min(numeros)
maior = max(numeros)
media = sum(numeros) / len(numeros)

# Exibir resultados

# Definir a largura das colunas
largura_col1 = 20
largura_col2 = 10

# Exibir cabeçalho
print(f"{'Descrição':<{largura_col1}}{'Valor':<{largura_col2}}")
print("-" * (largura_col1 + largura_col2))

# Exibir resultados
print(f"{'Menor número:':<{largura_col1}}{menor:<{largura_col2}}")
print(f"{'Maior número:':<{largura_col1}}{maior:<{largura_col2}}")
print(f"{'Média dos números:':<{largura_col1}}{media:<{largura_col2}.2f}")
```
