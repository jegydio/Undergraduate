## E06 - Programa para converter segundos em horas, minutos e segundos

```python
tempo_total = float(input('Digite o tempo em segundos: '))

horas = tempo_total // 3600
minutos = (tempo_total % 3600) // 60
segundos = tempo_total % 60

print(f"{tempo_total} segundos corresponde a {horas} horas, {minutos} minutos e {segundos} segundos")
```

```python
import datetime

tempo_total = float(input('Digite o tempo em segundos: '))

def convert(tempo_total): 
    return str(datetime.timedelta(seconds = tempo_total))  

print(f"{tempo_total} segundos corresponde a {convert(tempo_total)}")
```

## E07 - Consumidor e itens vendidos na loja

```python
preco_item_1 = float(input('Valor do ítem #1: R$ '))
preco_item_2 = float(input('Valor do ítem #2: R$ '))
preco_item_3 = float(input('Valor do ítem #3: R$ '))

subtotal = preco_item_1 + preco_item_2 + preco_item_3
impostos = 0.07*subtotal
total = subtotal + impostos

print(f"\n Subtotal = R$ {subtotal:.2f} \n Impostos = R$ {impostos:.2f} \n Total da compra = R$ {total:.2f} ")
```

## E08 - Cálculo de investimentos

image.png

Onde:

*   *P* é o valor presente (quantia que você precisa depositar hoje)
*   *F* é o valor futuro (quantia que você deseja sacar após 10 anos)
*   *r* é a taxa de jutos anual
*   n é o prazo de investimento (medido em anos)

Suponha que você queira depositar uma certa quantia em uma conta poupan~ça, deixando render pelos próximos 10 anos. Ao final desse prazo, você deseja ter um montante de $10.000. Quanto você precisa depositar hoje para que isso aconteça?



```python
F = float(input('Valor desejado de saque: R$ '))
r = float(input('Taxa de juros da aplicação [formato 0 = 0% e 1 = 100%]: '))
n = float(input('Prazo [em anos]: '))

P = F/(1+r)**n

print(f"\n Valor de depósito presente = R$ {P:.2f} ")
```
