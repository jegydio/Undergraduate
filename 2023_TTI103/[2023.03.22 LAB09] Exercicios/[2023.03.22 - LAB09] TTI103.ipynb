```python
# Vers√£o 01

import random # m√≥dulo random

N = int(input("Digite um n√∫mero de pontos para o m√©todo de Monte Carlo: "))
point_in = 0

for i in range(N) :
  x = random.uniform(-1,1) # n√∫mero real (float) entre -1 e 1
  y = random.uniform(-1,1) # n√∫mero real (float) entre -1 e 1
  if (x*x+y*y) <= 1 :
    point_in = point_in + 1
print("Pi = ", (4*point_in)/N)
```

```python
# Vers√£o 02

import random                          # Importa o m√≥dulo random para gerar n√∫meros aleat√≥rios
import matplotlib.pyplot as plt        # Importa a biblioteca matplotlib.pyplot para criar gr√°ficos
from matplotlib.patches import Circle  # Importa a fun√ß√£o Circle para criar um c√≠rculo

N = int(input("Digite um n√∫mero de pontos para o m√©todo de Monte Carlo: "))  
point_in = 0    # Inicializa a contagem de pontos dentro do c√≠rculo

x_in = []       # Lista para armazenar as coordenadas x dos pontos dentro do c√≠rculo
y_in = []       # Lista para armazenar as coordenadas y dos pontos dentro do c√≠rculo
x_out = []      # Lista para armazenar as coordenadas x dos pontos fora do c√≠rculo
y_out = []      # Lista para armazenar as coordenadas y dos pontos fora do c√≠rculo

# Gera N pontos e verifica se est√£o out ou fora do c√≠rculo
for i in range(N):
    x = random.uniform(-1, 1)   # Gera uma coordenada x aleat√≥ria entre -1 e 1
    y = random.uniform(-1, 1)   # Gera uma coordenada y aleat√≥ria entre -1 e 1
    if (x * x + y * y) <= 1:    # Verifica se o ponto (x, y) est√° dentro do c√≠rculo
        point_in = point_in + 1 # Incrementa a contagem de pontos dentro do c√≠rculo
        x_in.append(x)          # Adiciona a coordenada x √† lista x_in
        y_in.append(y)          # Adiciona a coordenada y √† lista y_in
    else:
        x_out.append(x)         # Adiciona a coordenada x √† lista x_out
        y_out.append(y)         # Adiciona a coordenada y √† lista y_out

print("Pi = ", (4 * point_in) / N)   # Calcula e exibe a estimativa de Pi

# Cria o gr√°fico de dispers√£o e o c√≠rculo
fig, ax = plt.subplots(figsize=(6, 6))                                        # Cria uma figura e um objeto de eixos com tamanho 6x6
circle = Circle((0, 0), 1, edgecolor='black', facecolor='none', linewidth=3)  # Cria um objeto Circle com centro (0,0), raio 1 e cor de borda preta
ax.add_patch(circle)                                                          # Adiciona o c√≠rculo ao objeto de eixos
ax.scatter(x_in, y_in, color='blue', label='Dentro do c√≠rculo')               # Plota os pontos dentro do c√≠rculo em azul
ax.scatter(x_out, y_out, color='red', label='Fora do c√≠rculo')                # Plota os pontos fora do c√≠rculo em vermelho
ax.legend()         # Exibe a legenda do gr√°fico
ax.set_xlim(-1, 1)  # Define o limite do eixo x entre -1 e 1
ax.set_ylim(-1, 1)  # Define o limite do eixo y entre -1 e 1
ax.set
```
