## Aplicações em OCR

https://ai-facets.org/tesseract-ocr-best-practices/

- Carregar imagem

```python
img_ocr = cv2.imread('/content/drive/MyDrive/Colab Notebooks/_fiap/_SML/2024/[ Semestre 2 ]/[ Aula 15 ]/aula2/IMG_OCR.jpg')
```

```python
plt.figure(figsize=(20,10))
plt.imshow(img_ocr)
```

- Girar imagem

```python
angulo = -90
img_rotated = imutils.rotate_bound(img_ocr, angle = angulo)
plt.imshow(img_rotated)
plt.show()
```

- Transformar em texto

```python
text = str(pt.image_to_string(img_rotated, lang='por', config='--psm 6'))
```

```python
text
```

```python
with open('texto_ocr.txt', 'w') as f:
    f.write(text)
```
