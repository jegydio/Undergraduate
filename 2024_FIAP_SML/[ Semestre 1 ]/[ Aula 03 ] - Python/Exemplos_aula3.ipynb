## Análises

### Adicionar/modificar colunas

```
df_dados_paises['ln_pop'] = df_dados_paises['populacao'].apply(lambda x: np.log(x))
```

Esse código realiza uma operação em um DataFrame do Pandas chamado `df_dados_paises`. Vamos destrinchar o que cada parte faz:

- **`df_dados_paises`** é o seu DataFrame, que presumivelmente contém dados sobre países. Um DataFrame é como uma tabela com linhas e colunas, onde cada coluna tem um nome.

- **`['ln_pop']`** indica que você está criando ou modificando uma coluna chamada `ln_pop` no DataFrame `df_dados_paises`. Se a coluna já existir, seus valores serão substituídos; se não, uma nova coluna será criada.

- **`df_dados_paises['populacao']`** acessa a coluna `populacao` do DataFrame, que, como o nome sugere, provavelmente contém dados sobre a população dos países.

- **`.apply(lambda x: np.log(x))`** é onde a mágica acontece:
  - `.apply()` é um método dos DataFrames do Pandas que aplica uma função ao longo de um eixo do DataFrame. Por padrão, ele aplica a função às colunas, mas você pode configurá-lo para aplicar às linhas também.
  - `lambda x: np.log(x)` é uma função anônima (lambda) que pega um valor `x` e retorna o logaritmo natural desse valor. Aqui, `np.log()` é uma função do Numpy que calcula o logaritmo natural. Cada elemento da coluna `populacao` é passado para essa função lambda como `x`, e o resultado é o logaritmo natural da população.
  
Portanto, o que essa linha de código faz é criar (ou atualizar) uma coluna chamada `ln_pop` no DataFrame `df_dados_paises`, onde cada valor é o logaritmo natural da respectiva população na coluna `populacao`. Isso é útil em análises estatísticas e econômicas, pois o logaritmo da população pode ser uma forma mais útil de visualizar e analisar dados, especialmente quando há grandes variações nos tamanhos das populações entre os países. Transformações logarítmicas ajudam a normalizar a distribuição dos dados e a reduzir a influência de outliers (valores atípicos).

### Sumarizar dados

```
df_dados_paises.agg(idh_total_medio = pd.NamedAgg(column = 'idh', aggfunc = sum))
```

O código que usa o método `.agg()` do Pandas, que é usado para aplicar uma ou mais operações de agregação a um DataFrame. Vamos entender cada parte desse comando:

- **`df_dados_paises`**: É o DataFrame que contém os seus dados. Suponho que ele inclua várias informações sobre países, entre elas o Índice de Desenvolvimento Humano (IDH) de cada um.

- **`.agg()`**: Este método é usado para aplicar funções de agregação ao DataFrame. Ele permite que você faça operações de resumo ou estatísticas em dados agrupados.

- **`idh_total_medio = pd.NamedAgg(column='idh', aggfunc=sum)`**: Dentro do método `.agg()`, você está especificando uma operação de agregação nomeada. Vamos decompor isso:

  - **`idh_total_medio`**: Este é o nome que você está dando ao resultado da operação de agregação. Em outras palavras, o resultado dessa agregação será rotulado como `idh_total_medio` no DataFrame resultante.

  - **`pd.NamedAgg`**: É uma funcionalidade do Pandas que permite nomear o resultado de uma operação de agregação. Isso é útil quando você quer especificar nomes claros para as suas colunas resultantes, especialmente quando está fazendo múltiplas agregações de uma vez.

  - **`column='idh'`**: Aqui você está especificando qual coluna do seu DataFrame você deseja agregar. Neste caso, você está se referindo à coluna `idh`, que presumivelmente contém os valores do Índice de Desenvolvimento Humano para diferentes países.

  - **`aggfunc=sum`**: Aqui você define qual função de agregação você quer usar. `sum` significa que você quer somar os valores. Portanto, você está somando todos os valores na coluna `idh`.

Portanto, ao usar `df_dados_paises.agg(idh_total_medio = pd.NamedAgg(column='idh', aggfunc=sum))`, você está somando todos os valores na coluna `idh` do seu DataFrame `df_dados_paises` e nomeando o resultado dessa soma como `idh_total_medio`. Essa operação pode ser útil para calcular a soma total do IDH de todos os países listados no seu DataFrame, por exemplo, como parte de uma análise mais ampla sobre desenvolvimento humano global.

### Pivot: extender os dados em chave e valor

```
df_dados_paises.pivot(index = 'ano', columns = 'pais', values = 'idh').head()
```

Este comando utiliza o método `.pivot()` do Pandas, seguido de `.head()`, para transformar o DataFrame `df_dados_paises` de modo a reorganizar os dados especificados. Vamos detalhar cada parte do comando:

1. **`df_dados_paises.pivot(index='ano', columns='pais', values='idh')`**:
   - **`pivot()`** é um método que permite reestruturar ou remodelar um DataFrame. Ele retorna um DataFrame reorganizado com base em três parâmetros: `index`, `columns`, e `values`.
   - **`index='ano'`**: Este parâmetro define a coluna do DataFrame original que se tornará o novo índice (ou seja, as linhas) do DataFrame reestruturado. Cada valor único na coluna `ano` se tornará uma linha distinta no novo DataFrame.
   - **`columns='pais'`**: Este parâmetro especifica qual coluna do DataFrame original será usada para criar as novas colunas no DataFrame reestruturado. Assim, cada valor único encontrado na coluna `pais` se tornará uma coluna separada no novo DataFrame.
   - **`values='idh'`**: Este parâmetro indica qual coluna do DataFrame original será usada para preencher os valores no novo DataFrame. Os valores da coluna `idh` serão redistribuídos no novo DataFrame conforme especificado pelos parâmetros `index` e `columns`.

2. **`.head()`**:
   - Este método é usado para visualizar as primeiras linhas de um DataFrame. Por padrão, `.head()` retorna as primeiras 5 linhas, mas você pode passar um número como argumento para alterar esse comportamento (por exemplo, `.head(10)` para as primeiras 10 linhas).

Portanto, o que este comando faz é reorganizar `df_dados_paises` para que cada linha represente um ano único (`index='ano'`), cada coluna represente um país único (`columns='pais'`), e as células contenham os valores do Índice de Desenvolvimento Humano (`values='idh'`) correspondentes a cada combinação de ano e país. O método `.head()` é então usado para visualizar as primeiras 5 linhas deste DataFrame reestruturado, oferecendo uma rápida visão geral de como os dados foram organizados.

Essa reorganização é útil para análises que exigem a comparação de valores ao longo do tempo para diferentes países, facilitando a visualização e análise de tendências ou padrões no Índice de Desenvolvimento Humano entre diversos países ao longo dos anos.

### Melt: selecionar colunas e agrupar

```
pd.melt(df_dados_paises, id_vars=['pais', 'ano'], var_name = 'indicadores', value_name = 'indices')
```

O código utiliza a função `pd.melt()` do Pandas, que é usada para transformar ou "derreter" o DataFrame de um formato largo para um formato longo. Especificamente, essa função pode ser usada para transformar um conjunto de colunas em um formato onde há apenas uma coluna que contém todos os valores e uma coluna que identifica esses valores. Vamos examinar cada parte do comando:

- **`pd.melt()`**: Esta é a função do Pandas usada para "derreter" o DataFrame. O primeiro argumento é o DataFrame que você deseja transformar.

- **`df_dados_paises`**: Este é o DataFrame de entrada que você está transformando. Presumivelmente, contém dados sobre diferentes países, incluindo várias colunas para diferentes indicadores por ano.

- **`id_vars=['pais', 'ano']`**: O argumento `id_vars` especifica as colunas que devem permanecer como estão, servindo como identificadores. No seu caso, `pais` e `ano` são as colunas que você deseja manter inalteradas, pois cada linha será única para cada combinação de país e ano.

- **`var_name='indicadores'`**: Este argumento permite renomear a coluna de variáveis (anteriormente nomes de colunas no formato largo) para `indicadores`. Ou seja, as colunas que não são `id_vars` e que serão "derretidas" terão seus nomes movidos para valores em uma nova coluna chamada `indicadores`.

- **`value_name='indices'`**: Este argumento especifica o nome da nova coluna que armazenará os valores correspondentes aos `indicadores`. Então, todos os valores que estavam nas colunas "derretidas" serão agora reunidos nesta coluna, chamada `indices`.

O resultado deste comando é um DataFrame no formato longo, onde você tem uma linha para cada combinação de país, ano, e indicador, com o respectivo valor desse indicador. Isso é particularmente útil para simplificar o DataFrame, tornando-o mais adequado para certas formas de análise estatística, visualizações ou quando se trabalha com ferramentas que esperam dados em um formato longo. Por exemplo, se originalmente você tinha colunas separadas para diferentes indicadores econômicos, demográficos, ou de saúde para cada país e ano, após o uso de `pd.melt()`, você terá esses indicadores todos em uma única coluna, com seus valores correspondentes em outra, facilitando a análise comparativa entre eles.

### Concatenar valores das colunas

```
df_dados_paises['pais'] + '_' + df_dados_paises['ano'].astype(str)
```

```
df_dados_paises['pais'].str.cat(df_dados_paises['ano'].astype(str), sep="_")
```

Este comando combina duas colunas do DataFrame `df_dados_paises`, especificamente as colunas `pais` e `ano`, em uma única coluna de strings, usando um separador específico. Vamos analisar cada parte:

1. **`df_dados_paises['pais']`**: Esta parte acessa a coluna `pais` do DataFrame `df_dados_paises`. Presume-se que contém os nomes dos países como strings.

2. **`.str`**: Este é um acessor que permite aplicar funções específicas de string vetorizadas às Series do Pandas. Basicamente, permite manipular os dados da coluna como se fossem strings.

3. **`.cat()`**: Este método é usado para concatenar strings. Você pode usá-lo para juntar strings de duas ou mais colunas.

4. **`df_dados_paises['ano'].astype(str)`**: Aqui, a coluna `ano` do DataFrame `df_dados_paises` é convertida para strings usando `.astype(str)`. Isso é necessário porque os anos provavelmente estão armazenados como inteiros, e você precisa deles como strings para poder concatená-los com os nomes dos países.

5. **`sep="_"`**: Este parâmetro especifica o separador a ser usado na concatenação. Neste caso, `_` é usado como separador entre o nome do país e o ano.

Portanto, o comando completo `df_dados_paises['pais'].str.cat(df_dados_paises['ano'].astype(str), sep="_")` cria uma Series no Pandas onde cada elemento é a concatenação do nome do país com o ano, separados por um sublinhado (`_`). Por exemplo, se tivermos um país "Brasil" no ano "2020", o resultado seria "Brasil_2020".

Este tipo de manipulação é útil quando você deseja criar identificadores únicos a partir de duas ou mais colunas ou quando precisa preparar os dados para análises específicas, visualizações ou exportação, onde uma representação combinada de várias colunas em uma única string é necessária.
