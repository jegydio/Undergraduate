```python
# versão 02

import random                     # módulo random

numero = random.randrange(1, 101) # número entre 1 e 100
palpites = 1
meu_palpite = int(input("Adivinhe meu número entre 1 e 100: "))

while meu_palpite != numero:
    if meu_palpite == 0:
      print("Ah. Você desistiu do jogo.")
      break
    else:
      palpites = palpites + 1
      if meu_palpite > numero:
        print(meu_palpite, "está acima.")
      elif meu_palpite < numero:
        print(meu_palpite, "está abaixo.")
      meu_palpite = int(input("tente novamente: "))

if meu_palpite != 0:
  print("\nÓtimo, você acertou em", palpites, "tentativas!")
```

```python
# Versão 01

import random # módulo random

N = int(input("Digite um número de pontos para o método de Monte Carlo: "))
point_in = 0

for i in range(N) :
  x = random.uniform(-1,1) # número real (float) entre -1 e 1
  y = random.uniform(-1,1) # número real (float) entre -1 e 1
  if (x*x+y*y) <= 1 :
    point_in = point_in + 1
print("Pi = ", (4*point_in)/N)
```

```python
# Versão 02

import random                          # Importa o módulo random para gerar números aleatórios
import matplotlib.pyplot as plt        # Importa a biblioteca matplotlib.pyplot para criar gráficos
from matplotlib.patches import Circle  # Importa a função Circle para criar um círculo

N = int(input("Digite um número de pontos para o método de Monte Carlo: "))  
point_in = 0    # Inicializa a contagem de pontos dentro do círculo

x_in = []       # Lista para armazenar as coordenadas x dos pontos dentro do círculo
y_in = []       # Lista para armazenar as coordenadas y dos pontos dentro do círculo
x_out = []      # Lista para armazenar as coordenadas x dos pontos fora do círculo
y_out = []      # Lista para armazenar as coordenadas y dos pontos fora do círculo

# Gera N pontos e verifica se estão out ou fora do círculo
for i in range(N):
    x = random.uniform(-1, 1)   # Gera uma coordenada x aleatória entre -1 e 1
    y = random.uniform(-1, 1)   # Gera uma coordenada y aleatória entre -1 e 1
    if (x * x + y * y) <= 1:    # Verifica se o ponto (x, y) está dentro do círculo
        point_in = point_in + 1 # Incrementa a contagem de pontos dentro do círculo
        x_in.append(x)          # Adiciona a coordenada x à lista x_in
        y_in.append(y)          # Adiciona a coordenada y à lista y_in
    else:
        x_out.append(x)         # Adiciona a coordenada x à lista x_out
        y_out.append(y)         # Adiciona a coordenada y à lista y_out

print("Pi = ", (4 * point_in) / N)   # Calcula e exibe a estimativa de Pi

# Cria o gráfico de dispersão e o círculo
fig, ax = plt.subplots(figsize=(6, 6))                                        # Cria uma figura e um objeto de eixos com tamanho 6x6
circle = Circle((0, 0), 1, edgecolor='black', facecolor='none', linewidth=3)  # Cria um objeto Circle com centro (0,0), raio 1 e cor de borda preta
ax.add_patch(circle)                                                          # Adiciona o círculo ao objeto de eixos
ax.scatter(x_in, y_in, color='blue', label='Dentro do círculo')               # Plota os pontos dentro do círculo em azul
ax.scatter(x_out, y_out, color='red', label='Fora do círculo')                # Plota os pontos fora do círculo em vermelho
ax.legend()         # Exibe a legenda do gráfico
ax.set_xlim(-1, 1)  # Define o limite do eixo x entre -1 e 1
ax.set_ylim(-1, 1)  # Define o limite do eixo y entre -1 e 1
ax.set
```
